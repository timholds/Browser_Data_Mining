```python
def url_shortener(df):
    if (df.url[0:3] == 'chr'):
        return df.url[71:]
    else:
        return df.url
        
df['url_short'] = df.apply(url_shortener, axis=1)
#df
```

```python
def date_time(df):
    df['time_usec'] = df['time_usec']/1000000
    return datetime.datetime.fromtimestamp(
        int(df['time_usec'])
    ).strftime('%Y-%m-%d %H:%M:%S')        
    
df['dates_and_times'] = df.apply(date_time, axis=1)
df
```
